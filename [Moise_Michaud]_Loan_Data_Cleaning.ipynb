```python
# Re-load the original DataFrame to ensure 'CCAvg' column exists, as 'df' was overwritten by a previous example cell.
import pandas as pd
file_name = 'bank_loan (3).csv' # This filename is based on the kernel state and previous uploads.
df = pd.read_csv(file_name)

print("\n--- Fixing 'CCAvg' Column Inconsistency ---")

def parse_ccavg(value):
    if isinstance(value, str):
        if '/' in value:
            try:
                numerator, denominator = map(float, value.split('/'))
                # Handle division by zero if it occurs, otherwise perform division
                return numerator / denominator if denominator != 0 else 0.0
            except ValueError:
                # If conversion to float fails, return NaN
                return pd.NA
        else:
            # If it's a string but no '/', try to convert directly to float
            try:
                return float(value)
            except ValueError:
                return pd.NA
    return value # Return original value if not a string

# Apply the parsing function to the 'CCAvg' column
df['CCAvg'] = df['CCAvg'].apply(parse_ccavg)

# Convert the column to numeric, coercing errors to NaN
df['CCAvg'] = pd.to_numeric(df['CCAvg'], errors='coerce')

# Check for any NaNs introduced during conversion and fill with median if desired
if df['CCAvg'].isnull().any():
    print("NaN values introduced in 'CCAvg' during conversion. Filling with median...")
    df['CCAvg'].fillna(df['CCAvg'].median(), inplace=True)

print("Current data type of 'CCAvg':", df['CCAvg'].dtype)
print("First five rows of 'CCAvg' after cleaning:")
print(df['CCAvg'].head())
```

```python
df.to_csv('bank_loan_cleaned.csv', index=False)
print("Cleaned DataFrame saved to 'bank_loan_cleaned.csv'")
```
