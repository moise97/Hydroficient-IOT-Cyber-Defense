```python
import json

# Define the path to the JSON file
json_file_path = '/content/EXTERN_sample.json'

# Load the JSON data
with open(json_file_path, 'r') as f:
    data = json.load(f)

print(f"Successfully loaded data from {json_file_path}")
# Display the first part of the loaded data to verify
print("First entry of the loaded data:")
print(json.dumps(data[0], indent=2))
```

### Task 2: Retrieve all order IDs along with their corresponding total amounts

Now, we'll iterate through the loaded data to extract each `order_id` and its `total` amount.

```python
order_totals = []

for record in data:
    customer_orders = record.get('orders', [])
    for order in customer_orders:
        order_id = order.get('order_id')
        total_amount = order.get('total')
        if order_id is not None and total_amount is not None:
            order_totals.append({'order_id': order_id, 'total_amount': total_amount})

# Display the retrieved order IDs and total amounts
print("Order IDs and their total amounts:")
for item in order_totals[:5]: # Displaying first 5 for brevity
    print(f"Order ID: {item['order_id']}, Total Amount: {item['total_amount']}")

# You can also display all of them if needed
# print(order_totals)
```

### Task 3: Normalize the JSON structure using Pandas to create a flat DataFrame

We will use the `pandas` library to flatten the nested JSON data. We need to iterate through each customer, then through each of their orders, and finally through each item in those orders to create a comprehensive, flat table.

```python
import pandas as pd

# Prepare a list to hold the flattened records
flattened_data = []

for customer_record in data:
    customer_id = customer_record.get('customer_id')
    customer_name = customer_record.get('name')

    for order in customer_record.get('orders', []):
        order_id = order.get('order_id')
        order_total = order.get('total')

        for item in order.get('items', []):
            flattened_data.append({
                'customer_id': customer_id,
                'customer_name': customer_name,
                'order_id': order_id,
                'product': item.get('product'),
                'quantity': item.get('quantity'),
                'total_per_item': item.get('price') * item.get('quantity'), # Calculate total per item for clarity
                'order_total': order_total # The total for the entire order
            })

# Create a DataFrame from the flattened data
df_normalized = pd.DataFrame(flattened_data)

print("Normalized DataFrame created successfully. Here are the first 5 rows:")
print(df_normalized.head())

print("\nDataFrame Info:")
df_normalized.info()
```

### Task 4: Save the processed DataFrame to a new JSON file

Now, we'll save the `df_normalized` DataFrame to a new JSON file. This file will be named `normalized_orders.json` and saved in the Colab environment.

```python
# Define the path for the new JSON file
output_json_file_path = '/content/normalized_orders.json'

# Save the DataFrame to a JSON file. Use 'records' orient to get a list of objects.
df_normalized.to_json(output_json_file_path, orient='records', indent=4)

print(f"DataFrame successfully saved to {output_json_file_path}")
print("You can find this file in the Colab file browser (left-hand sidebar, folder icon) and download it.")
```
